```python
import mongoengine
from mongoengine import Document
from mongoengine import DateTimeField
from mongoengine import DictField
from mongoengine import EmailField
from mongoengine import EmbeddedDocument
from mongoengine import EmbeddedDocumentField
from mongoengine import IntField
from mongoengine import ListField
from mongoengine import ReferenceField
from mongoengine import StringField
from mongoengine import URLField
from datetime import datetime


class DefaultOptions(object):
    
    '''default product options'''
    __default_cdr_ecv_options = {
        'sourcefile'      : False,           #delivers underlying raster product as part of order
        'source_metadata' : False,           #delivers underlying metadata
        'toa'             : False,           #deliver LEDAPS based top of atmosphere
        'thermal'         : False,           #deliver LEDAPS band 6
        'sr'              : False,           #deliver LEDAPS surface reflectance
        'sr_browse'       : False,           #generate and deliver surface reflectance browse image
        'sr_ndvi'         : False,           #deliver normalized difference vegetation
        'sr_ndmi'         : False,           #deliver normalized difference moisture
        'sr_nbr'          : False,           #deliver normalized burn ratio
        'sr_nbr2'         : False,           #deliver normalized burn ratio 2
        'sr_savi'         : False,           #deliver soil adjusted vegetation
        'sr_msavi'        : False,           #deliver modified soil adjusted vegetation
        'sr_evi'          : False,           #deliver enhanced vegetation
        'swe'             : False,           #deliver surface water extent
        'sca'             : False,           #deliver snow covered area
        'solr_index'      : False,           #deliver a solr search index record
        'cfmask'          : False            #deliver cfmask in a seperate file (normally delivered in sr)
    }

    
    '''default product options'''
    __default_lpvs_options = {}

    
    '''default projection options'''
    __default_projection_options = {
                                             #reprojection options
        'reproject'              : False,    #reproject all rasters (True/False)
        'target_projection'      : None,     #if 'reproject' which projection?
        'central_meridian'       : None,     #
        'false_easting'          : None,     #
        'false_northing'         : None,     #
        'origin_lat'             : None,     #
        'std_parallel_1'         : None,     #
        'std_parallel_2'         : None,     #
        'datum'                  : 'wgs84',  #
                                             #utm only options
        'utm_zone'               : None,     # 1 to 60
        'utm_north_south'        : None      # north or south
    }


    '''default subset and framing options'''
    __default_subset_options {        
        'image_extents'          : False,    #modify image extents (subset or frame)
        'minx'                   : None,     #
        'miny'                   : None,     #
        'maxx'                   : None,     #
        'maxy'                   : None      #    
    }


    '''default resize options'''
    __default_resize_options {         
        'resize'                 : False,    #resize output product pixel size (True/False)
        'pixel_size'             : None,     #if resize, how big (valid range 30 to 1000 meters)
        'pixel_size_units'       : None      #meters or dd.     
    }

    
    '''default options required for resampling'''
    __default_resample_options {      
        'resample_method'        : 'near'    #Must have this when reprojecting or resizing pixels
                                             #for ops that need it, how would user like to resample?
    }


    def get(self, system):
        '''returns default options for the named system'''
        if system == "cdr_ecv":
            o = self.__default_cdr_ecv_options
            o.update(self.__default_product_options)
            o.update(self.__default_projection_options)
            o.update(self.__default_subset_options)
            o.update(self.__default_resize_options)
            o.update(self.__default_resample_options)    
            return o
        elif system == "lpvs":
            return self.__default_lpvs_options
        else:
            return dict()
    
    
class EventLog(Document):       
    timestamp = DateTimeField(default=datetime.now(), db_field="ts")
    event_class= StringField(db_field="ec")
    name = StringField(db_field="n")
    entry = DictField(db_field="e")

class User(Document):
    email = EmailField(unique=True, db_field="em")
    registration_id = IntField(db_field="regid")
    events = ListField(ReferenceField(EventLog, reverse_delete_rule=mongoengine.NULLIFY), db_field="evt")

class Distribution(EmbeddedDocument):
    server = StringField(db_field="srvr")
    disk_location = StringField(db_field="dl")
    url = URLField(db_field="u")
    
class Checksum(EmbeddedDocument):
    value = StringField(db_field="val")
    bytes = StringField(db_field="b")
    disk_location = StringField(db_field="dl")
    url = URLField(db_field="u")
    
class Scene(EmbeddedDocument):
    name = StringField(db_field="n")
    events = ListField(ReferenceField(EventLog), db_field="evt")
    distribution = EmbeddedDocumentField(Distribution, db_field="dist")
    checksum = EmbeddedDocumentField(Checksum, db_field="chk")

class Order(Document):
    name = StringField(unique=True, db_field="n")
    system = StringField(choices=('cdr_ecv', 'lpvs'), db_field="sys")
    source = StringField(choices=('bulk_ordering', 'api', 'ee'), db_field="src")
    status = StringField(choices=('ordered', 'processing', 'complete'), db_field="stat")
    stage = StringField(choices=('hadoop', 'composite'), db_field="stg")
    scenes = ListField(EmbeddedDocumentField(Scene), db_field="scn")
    options = DictField(default=get_default_options(system), db_field="opt")
    events = ListField(ReferenceField(EventLog, reverse_delete_rule=mongoengine.NULLIFY), db_field="evt")
    note = StringField(db_field="nt")

    
```
